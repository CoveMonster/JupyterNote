```python
# 计算信息熵
def InforEntropy(data):
        var = []
        #infor_Entroy = []
        # 统计数据中不同的值的个数
        for item in data:        
            flag = True
            for value in var:
                if item == value['value']:
                    value['num'] += 1
                    flag = False
            if flag :
                new_item = {}
                new_item['value'] = item
                new_item['num'] = 1
                var.append(new_item)
        size = len(data)
        infor_ent = 0
            # 计算信息熵
        for item in var:
            infor_ent += -((item['num']/size)*(math.log(item['num']/size, 2)))
        return infor_ent

# data = ['是', '是', '是','是', '是', '是','是', '是', '否', '否', '否',  '否', '否', '否',  '否', '否', '否']
# InforEntropy(data)
        
```

```python
# 连续变量离散化 : 输入 一维列表，返回 最佳 二分点  和 最大化信息增益
def Discretization(x_arr, y):
    # 得到划分点集合 len(x_arr) - 1
    x_sort = []
    for x in x_arr:
        x_sort.append(x)
    x_sort.sort()
    y_infor_ent = InforEntropy(y)
    max_gain, optimal_seg_point = 0, 0    
    for i in range(len(x_sort)-1):
        seg_point = (x_sort[i] + x_sort[i+1])/2        
        y_small = []
        y_large = []
        for j in range(len(x_arr)):
            #print(y[j])
            if x_arr[j] >= seg_point:
                y_large.append(y[j])
                
            else:
                y_small.append(y[j])
        new_gain = y_infor_ent - (len(y_large)/len(y))*InforEntropy(y_large) - (len(y_small)/len(y))*InforEntropy(y_small)
        #print('{0:d}个划分点：{1:0.4f} ; 小集：{2}; 大集：{3} ；信息增益 {4:0.3f}'.format(i, seg_point, str(y_small), str(y_large), new_gain))
        #print('{0:d}个划分点 : {1:0.4f}\t信息增益 : {2:0.3f}'.format(i, seg_point, new_gain))
        if max_gain < new_gain:
            max_gain = new_gain
            optimal_seg_point = seg_point
    return optimal_seg_point, max_gain
    
#     
# var_7 = [0.697, 0.774, 0.634, 0.608, 0.556, 0.403, 0.481, 0.437, 0.666, 0.243, 0.245, 0.343, 0.639, 0.657, 0.36, 0.593, 0.719]  
# print(type(var_7))
# var_7.sort(reverse=False)
# print(var_7)
# y_0 = ['是', '是', '是','是', '是', '是','是', '是', '否', '否', '否',  '否', '否', '否',  '否', '否', '否']
# print(Discretization(var_7, y_0))

```
