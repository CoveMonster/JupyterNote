```python
# 计算信息熵
def InforEntropy(data):
        var = []
        #infor_Entroy = []
        # 统计数据中不同的值的个数
        for item in data:        
            flag = True
            for value in var:
                if item == value['value']:
                    value['num'] += 1
                    flag = False
            if flag :
                new_item = {}
                new_item['value'] = item
                new_item['num'] = 1
                var.append(new_item)
        size = len(data)
        infor_ent = 0
            # 计算信息熵
        for item in var:
            infor_ent += -((item['num']/size)*(math.log(item['num']/size, 2)))
        return infor_ent
```

```python
# 连续变量离散化 : 输入 一维列表，返回 最佳 二分点  和 最大化信息增益

def DiscretizationGain(x_arr, y):
    # 得到划分点集合 len(x_arr) - 1
    x_sort = []
    for x in x_arr:
        x_sort.append(x)
    x_sort.sort()
    y_infor_ent = InforEntropy(y)
    # 初始化信息增益 和 最优化二分点位置
    max_gain, optimal_seg_point = 0, 0    
    for i in range(len(x_sort)-1):
        seg_point = (x_sort[i] + x_sort[i+1])/2        
        y_small = []
        y_large = []
        for j in range(len(x_arr)):
            #print(y[j])
            if x_arr[j] >= seg_point:
                y_large.append(y[j])
                
            else:
                y_small.append(y[j])
        new_gain = y_infor_ent - (len(y_large)/len(y))*InforEntropy(y_large) - (len(y_small)/len(y))*InforEntropy(y_small)
        #print('{0:d}个划分点：{1:0.4f} ; 小集：{2}; 大集：{3} ；信息增益 {4:0.3f}'.format(i, seg_point, str(y_small), str(y_large), new_gain))
        #print('{0:d}个划分点 : {1:0.4f}\t信息增益 : {2:0.3f}'.format(i, seg_point, new_gain))
        if max_gain < new_gain:
            max_gain = new_gain
            optimal_seg_point = seg_point
    return  max_gain, optimal_seg_point
```

```python
# 计算信息增益 ： x_arr 表示选择的属性， y_arr 表示结论
def InforGain(x_sample,y_arr):
    if len(x_sample):     
        x_value = []
        for x in x_sample:
            flag = True
            for value in x_value:
                if value == x :
                    flag = False
                    break
            if flag:
                x_value.append(x)
        ent_sum = 0 # 初始化 
        for i in range(len(x_value)):
            new_seg = []
            for j in range(len(x_sample)):
                if x_sample[j] == x_value[i]:
                    new_seg.append(y_arr[j])
            ent_sum += ((len(new_seg)/len(y_arr))*(InforEntropy(new_seg)))
        return InforEntropy(y_arr) - ent_sum
```

```python
import numpy as np
# 数据处理
def pre_data(var_1, var_2, var_3, var_4, var_5, var_6, var_7, var_8, y_0):
    data = []
    for i in range(0,17):
        new_item = [var_1[i], var_2[i], var_3[i], var_4[i], var_5[i], var_6[i], var_7[i], var_8[i], y_0[i]]
        data.append(new_item)
    return data


# 缺失值处理
def missingValueProcessing(index_data, y):
    new_data ,new_y = [],[]
    for pos in range(len(index_data)):
        if index_data[pos] == 'Null':
            continue
        else:
            new_data.append(index_data[pos])
            new_y.append(y[pos])
    return new_data, new_y
```

```python
def getIndexAttributes(data, index):    
    x_sample = np.array(data)[:, index].tolist()
    #print(x_sample)
    x_value = []
    for x in x_sample:
        if x != 'Null':
            flag = True
            for value in x_value:
                if value == x :
                    flag = False
                    break
            if flag:
                x_value.append(x)
    return x_value


def divSet_Dis(data, y, index, optimal_seg_point):
    div_x , div_y = [], []
    x_small, x_large, y_small, y_large = [], [], [], []
    for row in range(len(data)):
        if float(data[row][index]) > optimal_seg_point:
            x_large.append(data[row])
            y_large.append(y[row])
        else:
            x_small.append(data[row])
            y_small.append(y[row])
    div_x.append(np.delete(x_large,index,axis = 1))
    div_x.append(np.delete(x_small,index,axis = 1))
    div_y.append(y_large)
    div_y.append(y_small)
    return div_x, div_y    

def divSet_Con(data, y, index, attributes):
        #print('attributes : ',attributes)
    data = np.array(data)
    div_x = []
    div_y = []
    for i in range(len(attributes)):
        new_x = []
        new_y = []
        for row in range(len(data)):
            if data[row][index] == attributes[i] or data[row][index] == 'Null':
                new_x.append(data[row])
                new_y.append(y[row])
        div_x.append(np.delete(new_x,index,axis = 1))
        div_y.append(new_y)
            #div_set.append(new_set)
    return div_x, div_y
```

```python
# 计算每一个属性与y的信息增益，选择最大做为当前分支的标准
def getGainInfor(data, y ,label):
    row, index = data.shape
    max_gain, feature_pos, select_feature, optimal_seg_point = 0, 0, '',0
    # 从1开始，0表示样本序号
    for i in range(1,index-1):
        # 过滤缺失值 
        now_x, now_y = missingValueProcessing(data[:,i],y)
        # 计算信息增益；分连续值和离散值处理
        now_gain = 0
        if  '密度' == label[i]  or  '含糖率' == label[i] :
            float_now_x = []
            try:
                for x in now_x:
                    float_now_x.append(float(x))
                    now_gain, optimal_seg_point = DiscretizationGain(float_now_x, y)
            except ValueError  as e:
                print('-'*100,'异常')
                print(now_x)  
                print('-'*100,'异常')
        else: now_gain = InforGain(now_x, now_y)
        if max_gain < now_gain:
            max_gain, feature_pos, select_feature = now_gain, i, label[i]
    return  max_gain, feature_pos, select_feature, optimal_seg_point
        
        
```

```python
#这里的data\label为array类型
def produceNode(data,y,label):
    if InforEntropy(y) == 0:
        print ('****************************** 分支结束***********************************')
        print('叶子节点拥有样本集合：', data[:,0],'特征属性', label, '\n')
#         print('剩下的特征属性：{0} \n'.format(label))
        print ('****************************** 分支结束***********************************')
    else:
        print ('非叶子节点拥有样本集合：', data[:,0])
        # 计算信息增益， 返回节点信息，
        max_gain, feature_pos, select_feature, optimal_seg_point = getGainInfor(data, y, label)
        # 分割样本集，生成新的label
        div_x, div_y, attributes = [], [], []
        if  '密度' == label[feature_pos]  or  '含糖率' == label[feature_pos] :
            div_x, div_y = divSet_Dis(data, y, feature_pos, optimal_seg_point)
        else:
            attributes = getIndexAttributes(data, feature_pos) # 得到特征的各个属性值
            div_x, div_y = divSet_Con(data, y, feature_pos, attributes)
        new_label = np.delete(label, feature_pos, axis = None)
        print('当前选择的特征：{0} ；属性值 ：{1} ；连续值切分点 ：{2} ；其信息增益为：{3} ；\n 剩下的特征属性：{4} \n'.format(select_feature,attributes,optimal_seg_point , max_gain, new_label)) 
        # 用新的样本集 跑新Node
        print('---------------------------------------------------, 兄弟的个数 ： ', len(div_x))
        for order in range(len(div_x)):            
            #print('----------当前选择的特征：{0} ；其信息增益为：{1} ；分支属性值为 {2} \n'.format(select_feature, max_gain, attributes[order]))
            produceNode(div_x[order], div_y[order], new_label)
            
        
label = ['序号', '色泽', '根蒂', '敲声', '纹理', '脐部', '触感', '密度', '含糖率']
data = pre_data(var_0, var_1, var_2, var_3, var_4, var_5, var_6, var_7, var_8)
y = ['是', '是', '是','是', '是', '是','是', '是', '否', '否', '否',  '否', '否', '否',  '否', '否', '否']
data, y, label = np.array(data), np.array(y), np.array(label)
produceNode(data, y, label)

        
            
    
    
```
