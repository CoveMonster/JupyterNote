```python
import math
# 计算信息熵
def InforEntropy(data):
        var = []
        #infor_Entroy = []
        # 统计数据中不同的值的个数
        for item in data:        
            flag = True
            for value in var:
                if item == value['value']:
                    value['num'] += 1
                    flag = False
            if flag :
                new_item = {}
                new_item['value'] = item
                new_item['num'] = 1
                var.append(new_item)
        size = len(data)
        infor_ent = 0
            # 计算信息熵
        for item in var:            
            infor_ent += -((item['num']/size)*(math.log(item['num']/size, 2)))
        return infor_ent

# 得到某一列的属性值
def getIndexAttributes(x):    
    x_value = []
    for x in x:
        flag = True
        for value in x_value:
            if value == x :
                flag = False
                break
        if flag:
            x_value.append(x)
    return x_value
```

```python
# 计算离散变量的信息增益率
def D_InforGainRate(x_arr, y):
    # 得到划分点集合 len(x_arr) - 1
    x_sort = []
    for x in x_arr:
        x_sort.append(x)
    x_sort.sort()
    y_infor_ent = InforEntropy(y)
    # 初始化信息增益 和 最优化二分点位置
    max_gain_rate, optimal_seg_point = 0, 0    
    for i in range(len(x_sort)-1):
        seg_point = (x_sort[i] + x_sort[i+1])/2        
        y_small = []
        y_large = []
        for j in range(len(x_arr)):
            #print(y[j])
            if x_arr[j] >= seg_point:
                y_large.append(y[j])
                
            else:
                y_small.append(y[j])
        percent_large = len(y_large)/len(y)
        percent_small = len(y_small)/len(y)
        gain_rate = y_infor_ent - (percent_large)*InforEntropy(y_large)- (percent_small)*InforEntropy(y_small)
        iv = -percent_large*math.log(percent_large, 2)-percent_small*math.log(percent_small, 2)
        new_gain_rate = gain_rate/iv
        #print('{0:d}个划分点：{1:0.4f} ; 小集：{2}; 大集：{3} ；信息增益 {4:0.3f}'.format(i, seg_point, str(y_small), str(y_large), new_gain))
        #print('{0:d}个划分点 : {1:0.4f}\t信息增益 : {2:0.3f}'.format(i, seg_point, new_gain))
        if max_gain_rate < new_gain_rate:
            max_gain_rate = new_gain_rate
            optimal_seg_point = seg_point
    return  max_gain_rate, optimal_seg_point
```

```python
# 计算连续变量的信息增益率
def C_InforGainRate(x, y):
    x_value = getIndexAttributes(x)
    iv, ent_sum = 0, 0
    for i in range(len(x_value)):
        new_seg = []
        for j in range(len(x)):
            if x[j] == x_value[i]:
                new_seg.append(y[j])
        percent = len(new_seg)/len(y)
        ent_sum += percent*(InforEntropy(new_seg))
        iv += (-percent*math.log(percent, 2))
    return (InforEntropy(y) - ent_sum)/iv 
        
```

```python
def getInforGainRate(data,y,label):
    all_InforGainRate = []
    row, index = data.shape
    max_gain_rate, feature_pos, select_feature, optimal_seg_point = 0, 0, '',0
    for i in range(1, index - 1):
        new_label_gainRate = {}
        new_gain_rate = 0
        if label[i] == '密度' or label[i] == '含糖率':
            
            new_x = []
            for x in data[:,i]:
                new_x.append(float(x))
            new_gain_rate, optimal_seg_point = D_InforGainRate(new_x, y)
        else:
            new_gain_rate = C_InforGainRate(data[:,i], y)
        new_label_gainRate['label'] = label[i]
        new_label_gainRate['gain_rate'] = new_gain_rate
        all_InforGainRate.append(new_label_gainRate)
        if max_gain_rate < new_gain_rate:
            max_gain_rate, feature_pos, select_feature = new_gain_rate, i, label[i]
    print ('-'*20, '第{0}次计算信息增益率'.format(9-len(label)+1), '-'*20)
    print(all_InforGainRate)   
    return max_gain_rate, feature_pos, select_feature, optimal_seg_point
```

```python
def divSet_Dis(data, y, index, optimal_seg_point):
    div_x , div_y = [], []
    x_small, x_large, y_small, y_large = [], [], [], []
    for row in range(len(data)):
        if float(data[row][index]) > optimal_seg_point:
            x_large.append(data[row])
            y_large.append(y[row])
        else:
            x_small.append(data[row])
            y_small.append(y[row])
    div_x.append(np.delete(x_large,index,axis = 1))
    div_x.append(np.delete(x_small,index,axis = 1))
    div_y.append(y_large)
    div_y.append(y_small)
    return div_x, div_y    

def divSet_Con(data, y, index, attributes):
        #print('attributes : ',attributes)
    data = np.array(data)
    div_x = []
    div_y = []
    for i in range(len(attributes)):
        new_x = []
        new_y = []
        for row in range(len(data)):
            if data[row][index] == attributes[i]:
                new_x.append(data[row])
                new_y.append(y[row])
        div_x.append(np.delete(new_x,index,axis = 1))
        div_y.append(new_y)
            #div_set.append(new_set)
    return div_x, div_y

#这里的data\label为array类型
def produceNode(data,y,label):
    if InforEntropy(y) == 0:
        print ('****************************** 分支结束***********************************')
        print('叶子节点拥有样本集合：', data[:,0],'特征属性', label, '\n')
        print ('****************************** 分支结束***********************************')
    else:
        print ('非叶子节点拥有样本集合：', data[:,0])
        # 计算信息增益， 返回节点信息，
        max_gain, feature_pos, select_feature, optimal_seg_point = getInforGainRate(data, y, label)
        # 分割样本集，生成新的label
        div_x, div_y, attributes = [], [], []
        if  '密度' == label[feature_pos]  or  '含糖率' == label[feature_pos] :
            div_x, div_y = divSet_Dis(data, y, feature_pos, optimal_seg_point)
        else:
            attributes = getIndexAttributes(data[:,feature_pos]) # 得到特征的各个属性值
            div_x, div_y = divSet_Con(data, y, feature_pos, attributes)
        new_label = np.delete(label, feature_pos, axis = None)
        print('当前选择的特征：{0} ；属性值 ：{1} ；连续值切分点 ：{2} ；其信息增益率为：{3} ；\n '.format(select_feature,attributes,optimal_seg_point , max_gain)) 
        # 用新的样本集 跑新Node
        print('---------------------------------------------------, 兄弟的个数 ： ', len(div_x))
        for order in range(len(div_x)):            
            #print('----------当前选择的特征：{0} ；其信息增益为：{1} ；分支属性值为 {2} \n'.format(select_feature, max_gain, attributes[order]))
            produceNode(div_x[order], div_y[order], new_label)
            
```
